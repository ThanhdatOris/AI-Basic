```python
# @title Định dạng ngẫu nhiên cho ký tự trong chuỗi
import random

def random_format(text):
    formatted_text = ""
    for char in text:
        style = random.choice(['bold', 'italic', 'underline', 'uppercase', 'normal'])
        if style == 'bold':
            formatted_text += f"**{char}**"
        elif style == 'italic':
            formatted_text += f"*{char}*"
        elif style == 'underline':
            formatted_text += f"__{char}__"
        elif style == 'uppercase':
            formatted_text += char.upper()
        else:
            formatted_text += char
    return formatted_text

# Ví dụ sử dụng
input_text = "Xin chao cac ban"
print(random_format(input_text))
```

```python
# @title BFS-DFS AI-Gen
# Thuật toán tìm kiếm theo chiều ngang (BFS) và tìm kiếm theo chiều sâu (DFS)
# Trình bày dạng menu console

from collections import deque

# Định nghĩa đồ thị dưới dạng danh sách kề
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Hàm tìm kiếm theo chiều ngang (BFS)
def bfs(graph, start):
    visited = set()  # Tập hợp các đỉnh đã thăm
    queue = deque([start])  # Hàng đợi để lưu trữ các đỉnh cần thăm

    while queue:
        vertex = queue.popleft()  # Lấy đỉnh đầu tiên trong hàng đợi
        if vertex not in visited:
            print(vertex, end=" ")  # In đỉnh ra màn hình
            visited.add(vertex)  # Đánh dấu đỉnh đã thăm
            queue.extend(graph[vertex])  # Thêm các đỉnh kề vào hàng đợi

# Hàm tìm kiếm theo chiều sâu (DFS)
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()  # Tập hợp các đỉnh đã thăm

    print(start, end=" ")  # In đỉnh ra màn hình
    visited.add(start)  # Đánh dấu đỉnh đã thăm

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)  # Đệ quy thăm các đỉnh kề

# Hàm hiển thị menu
def menu():
    while True:
        print("\nChọn chế độ nhập đỉnh bắt đầu:")
        print("1. Nhập đỉnh bắt đầu thủ công")
        print("2. Sử dụng đỉnh bắt đầu cố định (A)")
        print("3. Thoát")
        mode_choice = input("Lựa chọn của bạn: ")

        if mode_choice == '1':
            start = input("Nhập đỉnh bắt đầu: ")
        elif mode_choice == '2':
            start = 'A'
        elif mode_choice == '3':
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
            continue

        print("\nChọn thuật toán tìm kiếm:")
        print("1. Tìm kiếm theo chiều ngang (BFS)")
        print("2. Tìm kiếm theo chiều sâu (DFS)")
        print("3. Thoát")
        algo_choice = input("Lựa chọn của bạn: ")

        if algo_choice == '1':
            print(f"Kết quả tìm kiếm theo chiều ngang (BFS) từ đỉnh {start}:")
            bfs(graph, start)
        elif algo_choice == '2':
            print(f"Kết quả tìm kiếm theo chiều sâu (DFS) từ đỉnh {start}:")
            dfs(graph, start)
        elif algo_choice == '3':
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")

# Chạy chương trình
menu()
```

# Best First Search - Uniform Cost Search

```python
# @title AI-Gen
import heapq

class Graph:
    def __init__(self):
        self.edges = {}
        self.weights = {}

    def add_edge(self, from_node, to_node, weight):
        if from_node not in self.edges:
            self.edges[from_node] = []
        self.edges[from_node].append(to_node)
        self.weights[(from_node, to_node)] = weight

def best_first_search(graph, start, goal, heuristic):
    open_list = []
    heapq.heappush(open_list, (heuristic(start, goal), start))
    closed_list = set()
    came_from = {}

    while open_list:
        _, current = heapq.heappop(open_list)
        if current == goal:
            return reconstruct_path(came_from, start, goal)

        closed_list.add(current)
        for neighbor in graph.edges.get(current, []):
            if neighbor not in closed_list:
                came_from[neighbor] = current
                heapq.heappush(open_list, (heuristic(neighbor, goal), neighbor))

    return None

def uniform_cost_search(graph, start, goal):
    open_list = []
    heapq.heappush(open_list, (0, start))
    closed_list = set()
    came_from = {}
    cost_so_far = {start: 0}

    while open_list:
        current_cost, current = heapq.heappop(open_list)
        if current == goal:
            return reconstruct_path(came_from, start, goal)

        closed_list.add(current)
        for neighbor in graph.edges.get(current, []):
            new_cost = current_cost + graph.weights[(current, neighbor)]
            if neighbor not in closed_list or new_cost < cost_so_far.get(neighbor, float('inf')):
                cost_so_far[neighbor] = new_cost
                came_from[neighbor] = current
                heapq.heappush(open_list, (new_cost, neighbor))

    return None

def reconstruct_path(came_from, start, goal):
    path = []
    current = goal
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()
    return path

def heuristic(node, goal):
    # Example heuristic function (Manhattan distance for grid)
    return abs(node[0] - goal[0]) + abs(node[1] - goal[1])

# Example usage
graph = Graph()
edges = [
    ('A', 'B', 1), ('A', 'C', 4), ('B', 'C', 2), ('B', 'D', 5),
    ('C', 'D', 1), ('D', 'E', 3)
]
for edge in edges:
    graph.add_edge(*edge)

start, goal = 'A', 'E'
print("Best First Search Path:", best_first_search(graph, start, goal, heuristic))
print("Uniform Cost Search Path:", uniform_cost_search(graph, start, goal))
```
